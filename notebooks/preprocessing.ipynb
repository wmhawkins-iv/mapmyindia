```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline

def filecheck(dir):
    dir_size = 0
    filelist = os.listdir(dir)
    filelist.sort()
    print(dir)
    for i,name in enumerate(filelist):
        dir_size += os.path.getsize(os.path.join(dir, name))
    print("{:.1f} MB of {} files".format(dir_size/1024/1024, i))
    print("sample files:")
    print("\n".join(filelist[:5]) + "\n")

dirs = ["../data/train","../data/test"]

for d in dirs:
    filecheck(d)
```

```python
train_data = pd.read_csv("../data/train.csv")

train_data.head()
```

```python
for i in range(5):
    print('Cropped Image', i, ':', train_data.iloc[i]["Label"])
    img_name = '../data/train/' + train_data.iloc[i]["Img_Name"]
    x = train_data.iloc[i]["Left"]
    y = train_data.iloc[i]["Top"]
    w = train_data.iloc[i]["Width"]
    h = train_data.iloc[i]["Height"]
    
    img = cv2.imread(img_name)
    crop_img = img[y:y+h, x:x+w]
    plt.imshow(crop_img)
    plt.show()
```
